```python
import seaborn as sns

import matplotlib.pyplot as plt

# Display basic statistics for each dataframe
print("Customer DataFrame Statistics:")
print(customer.describe(include='all'))
print("\nFilm DataFrame Statistics:")
print(film.describe(include='all'))
print("\nInventory DataFrame Statistics:")
print(inventory.describe(include='all'))
print("\nRental DataFrame Statistics:")
print(rental.describe(include='all'))

# Check for missing values
print("\nMissing Values in Customer DataFrame:")
print(customer.isnull().sum())
print("\nMissing Values in Film DataFrame:")
print(film.isnull().sum())
print("\nMissing Values in Inventory DataFrame:")
print(inventory.isnull().sum())
print("\nMissing Values in Rental DataFrame:")
print(rental.isnull().sum())

# Visualize the distribution of customer ages
plt.figure(figsize=(10, 6))
sns.histplot(customer['create_date'].dt.year, bins=20, kde=True)
plt.title('Distribution of Customer Creation Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()

# Visualize the distribution of film rental rates
plt.figure(figsize=(10, 6))
sns.histplot(film['rental_rate'], bins=20, kde=True)
plt.title('Distribution of Film Rental Rates')
plt.xlabel('Rental Rate')
plt.ylabel('Frequency')
plt.show()

# Visualize the number of films per rating
plt.figure(figsize=(10, 6))
sns.countplot(data=film, x='rating', order=film['rating'].value_counts().index)
plt.title('Number of Films per Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Visualize the number of rentals per month
rental['rental_month'] = rental['rental_date'].dt.to_period('M')
plt.figure(figsize=(14, 7))
rental['rental_month'].value_counts().sort_index().plot(kind='bar')
plt.title('Number of Rentals per Month')
plt.xlabel('Month')
plt.ylabel('Number of Rentals')
plt.show()
```
